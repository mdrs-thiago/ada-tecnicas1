## Indexação

Como acessamos diferentes índices em pandas? Até o momento, aprendemos apenas a visualizar algumas colunas, basicamente. 

Hoje veremos duas estratégias:

- `iloc`

- `loc`

Observação! Os dois paradigmas possuem a mesma estratégia para acesso: primeiro linha, depois coluna. 

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/mdrs-thiago/uerj-topicos-a/main/datasets/titanic/class_titanic.csv')
```

```python
df
```

`iloc` acessa as linhas e colunas por meio de seus índices numéricos. 

Exemplo:

`df.iloc[3:5, 5]`

```python
df.iloc[3:5, 5]
```

Podemos usar o `label` para fazer o acesso. Por exemplo:

`df.loc[5, 'label']`

```python
df.loc[3:5, 'SibSp']
```

E quando devemos usar `loc` ou `iloc`? Na verdade, depende...

## Manipulação de `DataFrame`

Os índices de uma tabela são, por default, ordenados numericamente. Mas não precisa ser **sempre** assim. E se quisermos modificar o índice? 

Deste modo, usamos o método `set_index`. 

Exemplo:

`df.set_index('date')`

```python
new_df = df.set_index('Ticket')

new_df.head()
```

```python
new_df.loc['113803']
```

### Seleção condicional

Usando `numpy`, vimos que podemos fazer comparações com diferentes operadores. O resultado da comparação é, neste caso, uma máscara contendo `True` ou `False`, quando a condição é satisfeita ou não, respectivamente. Será que isso pode ser feito usando `pandas`?

```python
df['Embarked']
```

```python
df['Embarked'] == 'S'
```

E como selecionamos? Temos duas formas:

- usando `loc`.

- usando índices normais (sem `loc`).

```python
df[df['Embarked'] == 'S']
```

```python
df.loc[df['Embarked'] == 'S']
```

`pandas` também tem algumas funções para realizar comparações. Neste caso, vamos citar:

- `isin`

- `isnull` (que é igual ao `isna`)

```python
df.isna()
```
